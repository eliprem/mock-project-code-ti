# Demographic Breakdowns

```python
def tobacco_charts(breakout_cat, yr, cat_order=None):
    tobacco_breakout = tobacco_df[((tobacco_df['QuestionID'] == '_tobaccoSTS') | (tobacco_df['QuestionID'] == 'tobaccoNOW1') | (tobacco_df['QuestionID'] == 'tobaccoNOW2')) & (tobacco_df['Break_Out_Category'] == breakout_cat)]
    tobacco_breakout = tobacco_breakout[['Year', 'Locationabbr', 'Question', 'Response', 'Break_Out', 'Break_Out_Category', 'Sample_Size', 'Data_value', 'QuestionID', 'ResponseID']]

    grouped_tobacco = tobacco_breakout.groupby(['Year', 'Response', 'Break_Out'])[['Sample_Size']].sum().reset_index()

    # TOTAL RESPONDENT CHART
    grouped_tobacco = grouped_tobacco[grouped_tobacco['Year'] == yr]
    # print(cat_order)
    # print(grouped_tobacco['Break_Out'])
    if(cat_order is not None):
        grouped_tobacco['Break_Out'] = pd.Categorical(grouped_tobacco['Break_Out'], categories=cat_order, ordered=True)
        sorted_tobacco = grouped_tobacco.sort_values('Break_Out')
        df_pivot = sorted_tobacco.pivot(index='Break_Out', columns='Response', values='Sample_Size').fillna(0)
    else:
        df_pivot = grouped_tobacco.pivot(index='Break_Out', columns='Response', values='Sample_Size').fillna(0)

    df_pivot.plot(kind='bar', stacked=True, figsize=(8, 6))
    plt.title(f'{breakout_cat} Breakdown of US Tobacco Usage Trends in {yr}')
    plt.xlabel(f'{breakout_cat}')
    plt.ylabel('Respondents')
    plt.legend(title='Response', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
    
    #PROPORTIONAL CHART
    df_pivot_normalized = df_pivot.div(df_pivot.sum(axis=1), axis=0) * 100

    df_pivot_normalized.plot(kind='bar', stacked=True, figsize=(8, 6))
    plt.title(f'PROPORTIONAL {breakout_cat} Breakdown of US Tobacco Usage Trends in {yr}')
    plt.xlabel(f'{breakout_cat}')
    plt.ylabel('Percentage of Respondents')
    plt.legend(title='Response', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

# Making Code to get BRFSS in right form a function (yay!)

```python
#make optional breakout column to be returned for demographic 

def brfss_extract(question_id, breakout):
    question_df = agg_df[agg_df['QuestionID'] == question_id]
    breakout_df = question_df[question_df['Break_Out_Category'] == breakout]

    if breakout == 'Overall':
        kept_cols = ['Year', 'Locationabbr', 'Question', 'Response', 'Data_value', 'QuestionID', 'ResponseID']
        sorted_df = breakout_df[kept_cols].sort_values(by=['Year', 'Locationabbr', 'ResponseID'])
    else:
        kept_cols = ['Year', 'Locationabbr', 'Question', 'Response', 'Data_value', 'Break_Out', 'QuestionID', 'ResponseID']
        sorted_df = breakout_df[kept_cols].sort_values(by=['Year', 'Locationabbr', 'ResponseID', 'Break_Out'])

    output_df = sorted_df.reset_index().drop(columns=['index'])
    
    return output_df
```

```python
smoker_simple_df = brfss_extract('_SMOKER3', 'Overall')

#Removing VI, because too many missing years
smoker_simple_df = smoker_simple_df[smoker_simple_df['Locationabbr'] != 'VI']

#NJ missing 2019, taking midpoint of 2018 and 2020
new_rows_nj = []
for responseID in ['RESP068', 'RESP069', 'RESP070', 'RESP071']:
    value_2018 = smoker_simple_df[(smoker_simple_df['Locationabbr'] == 'NJ') & (smoker_simple_df['Year'] == 2018) & (smoker_simple_df['ResponseID'] == responseID)]['Data_value'].values[0]
    value_2020 = smoker_simple_df[(smoker_simple_df['Locationabbr'] == 'NJ') & (smoker_simple_df['Year'] == 2020) & (smoker_simple_df['ResponseID'] == responseID)]['Data_value'].values[0]
    average_value = (value_2018 + value_2020) / 2
    new_row = {
        'Year': 2019,
        'Locationabbr': 'NJ',
        'Question': 'Four Level Smoking Status (variable calculated from one or more BRFSS questions)',
        'Response': 'Smoke everyday' if responseID == 'RESP068' else 'Smoke some days' if responseID == 'RESP069' else 'Former smoker' if responseID == 'RESP070' else 'Never smoked',
        'Data_value': average_value,
        'QuestionID': '_SMOKER3',
        'ResponseID': responseID,
    }
    new_rows_nj.append(new_row)

new_rows_nj_df = pd.DataFrame(new_rows_nj)
smoker_simple_df = pd.concat([smoker_simple_df, new_rows_nj_df], ignore_index=True)

#FL missing 2021, taking midpoint of 2020 and 2022
new_rows_fl = []
for responseID in ['RESP068', 'RESP069', 'RESP070', 'RESP071']:
    value_2020 = smoker_simple_df[(smoker_simple_df['Locationabbr'] == 'FL') & (smoker_simple_df['Year'] == 2020) & (smoker_simple_df['ResponseID'] == responseID)]['Data_value'].values[0]
    value_2022 = smoker_simple_df[(smoker_simple_df['Locationabbr'] == 'FL') & (smoker_simple_df['Year'] == 2022) & (smoker_simple_df['ResponseID'] == responseID)]['Data_value'].values[0]
    average_value = (value_2020 + value_2022) / 2
    new_row = {
        'Year': 2021,
        'Locationabbr': 'FL',
        'Question': 'Four Level Smoking Status (variable calculated from one or more BRFSS questions)',
        'Response': 'Smoke everyday' if responseID == 'RESP068' else 'Smoke some days' if responseID == 'RESP069' else 'Former smoker' if responseID == 'RESP070' else 'Never smoked',
        'Data_value': average_value,
        'QuestionID': '_SMOKER3',
        'ResponseID': responseID,
    }
    new_rows_fl.append(new_row)

new_rows_fl_df = pd.DataFrame(new_rows_fl)
smoker_simple_df = pd.concat([smoker_simple_df, new_rows_fl_df], ignore_index=True)

smoker_simple_df = smoker_simple_df.sort_values(by=['Year', 'Locationabbr', 'ResponseID'])

smoker_simple_df.head()
```

```python
#Null Check
nan_rows = smoker_simple_df[smoker_simple_df['Data_value'].isnull()]
nan_rows
```

```python
#smoker_simple_df = brfss_extract('_SMOKER3', 'Overall')

# Separate the data into smokes and doesn't smoke
smokes = smoker_simple_df[smoker_simple_df['ResponseID'].isin(['RESP068', 'RESP069'])] \
    .groupby(['Year', 'Locationabbr'])['Data_value'] \
    .sum() \
    .reset_index() \
    .rename(columns={'Data_value': 'smoke_pct'})

doesnt_smoke = smoker_simple_df[smoker_simple_df['ResponseID'].isin(['RESP070', 'RESP071'])] \
    .groupby(['Year', 'Locationabbr'])['Data_value'] \
    .sum() \
    .reset_index() \
    .rename(columns={'Data_value': 'dont_smoke_pct'})

# Merge the two data frames together
smoker_by_state_2011_2022 = pd.merge(smokes, doesnt_smoke, on=['Year', 'Locationabbr'], how='left')
smoker_by_state_2011_2022 = smoker_by_state_2011_2022.rename(columns={'Locationabbr': 'State'})
smoker_by_state_2011_2022.head()
```

```python
def brfss_extract(question_id, breakout):
    question_df = agg_df[agg_df['QuestionID'] == question_id]
    breakout_df = question_df[question_df['Break_Out_Category'] == breakout]

    if breakout == 'Overall':
        kept_cols = ['Year', 'Locationabbr', 'Question', 'Response', 'Data_value', 'QuestionID', 'ResponseID']
        sorted_df = breakout_df[kept_cols].sort_values(by=['Year', 'Locationabbr', 'ResponseID'])
    else:
        kept_cols = ['Year', 'Locationabbr', 'Question', 'Response', 'Data_value', 'Break_Out', 'QuestionID', 'ResponseID']
        sorted_df = breakout_df[kept_cols].sort_values(by=['Year', 'Locationabbr', 'ResponseID', 'Break_Out'])

    # Fill missing values if there's exactly one missing value per group
    def fill_missing_values(group):
        if group['Data_value'].isnull().sum() == 1:
            missing_index = group['Data_value'].isnull()
            group.loc[missing_index, 'Data_value'] = 100 - group['Data_value'].sum()
        return group

    if breakout == 'Overall':
        grouped_df = sorted_df.groupby(['Year', 'Locationabbr'])
    else:
        grouped_df = sorted_df.groupby(['Year', 'Locationabbr', 'Break_Out'])
    
    filled_df = grouped_df.apply(fill_missing_values)

    output_df = filled_df.reset_index(drop=True)
    
    return output_df
```

```python
smoker_age_df = brfss_extract('_SMOKER3', 'Age Group')
smoker_age_df
```

```python
#Null Check
nan_rows = smoker_age_df[smoker_age_df['Data_value'].isnull()]
nan_rows
```

```python
# Separate the data into smokes and doesn't smoke
def calculate_smoke_pct(df):
    if df['Data_value'].isnull().any():
        return pd.Series({'smoke_pct': None, 'dont_smoke_pct': None})
    smokes_age = df[df['ResponseID'].isin(['RESP068', 'RESP069'])]['Data_value'].sum()
    doesnt_smoke_age = df[df['ResponseID'].isin(['RESP070', 'RESP071'])]['Data_value'].sum()
    return pd.Series({'smoke_pct': smokes_age, 'dont_smoke_pct': doesnt_smoke_age})


smoker_by_state_by_age = smoker_age_df.groupby(['Year', 'Locationabbr', 'Break_Out']).apply(calculate_smoke_pct).reset_index()

smoker_by_state_by_age_2011_2022 = smoker_by_state_by_age.rename(columns={'Locationabbr': 'State'})
smoker_by_state_by_age_2011_2022.head()
```

```python
csv_file_path = 'smoker_by_state_by_age_2011_2022.csv'
smoker_by_state_by_age_2011_2022.to_csv(csv_file_path, index=False)
```
