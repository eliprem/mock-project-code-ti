Making a function for the charts

```python
def e_cig_charts(breakout_cat, yr, cat_order=None):
    e_cig_breakout = e_cig[((e_cig['QuestionID'] == '_ECIGSTS') | (e_cig['QuestionID'] == 'ECIGNOW1') | (e_cig['QuestionID'] == 'ECIGNOW2')) & (e_cig['Break_Out_Category'] == breakout_cat)]
    e_cig_breakout = e_cig_breakout[['Year', 'Locationabbr', 'Question', 'Response', 'Break_Out', 'Break_Out_Category', 'Sample_Size', 'Data_value', 'QuestionID', 'ResponseID']]

    grouped_ecig = e_cig_breakout.groupby(['Year', 'Response', 'Break_Out'])[['Sample_Size']].sum().reset_index()

    # TOTAL RESPONDENT CHART
    grouped_ecig = grouped_ecig[grouped_ecig['Year'] == yr]
    # print(cat_order)
    # print(grouped_ecig['Break_Out'])
    if(cat_order is not None):
        grouped_ecig['Break_Out'] = pd.Categorical(grouped_ecig['Break_Out'], categories=cat_order, ordered=True)
        sorted_ecig = grouped_ecig.sort_values('Break_Out')
        df_pivot = sorted_ecig.pivot(index='Break_Out', columns='Response', values='Sample_Size').fillna(0)
    else:
        df_pivot = grouped_ecig.pivot(index='Break_Out', columns='Response', values='Sample_Size').fillna(0)

    df_pivot.plot(kind='bar', stacked=True, figsize=(8, 6))
    plt.title(f'{breakout_cat} Breakdown of US ECig Usage Trends in {yr}')
    plt.xlabel(f'{breakout_cat}')
    plt.ylabel('Respondents')
    plt.legend(title='Response', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
    
    #PROPORTIONAL CHART
    df_pivot_normalized = df_pivot.div(df_pivot.sum(axis=1), axis=0) * 100

    df_pivot_normalized.plot(kind='bar', stacked=True, figsize=(8, 6))
    plt.title(f'PROPORTIONAL {breakout_cat} Breakdown of US ECig Usage Trends in {yr}')
    plt.xlabel(f'{breakout_cat}')
    plt.ylabel('Percentage of Respondents')
    plt.legend(title='Response', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

# Tobacco Charts

```python
tobacco = agg_df[agg_df['Class'] == 'Tobacco Use']
tobacco['Sample_Size'] = tobacco['Sample_Size'].str.replace(',', '').astype('float64')
tobacco['Sample_Size'] = tobacco['Sample_Size'].astype('float64')
#e_cig_income.dtypes
tobacco.head()
```

```python
tobacco_clean = tobacco[(tobacco['QuestionID'] == '_SMOKER3')]
tobacco_clean = tobacco_clean[['Year', 'Locationabbr', 'Question', 'Response', 'Break_Out', 'Break_Out_Category', 'Sample_Size', 'Data_value', 'QuestionID', 'ResponseID']]
tobacco_clean.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def tobacco_charts(breakout_cat, yr, cat_order=None, response_order=None, colors=None):
    tobacco_breakout = tobacco[(tobacco['QuestionID'] == '_SMOKER3') & (tobacco['Break_Out_Category'] == breakout_cat)]
    tobacco_breakout = tobacco_breakout[['Year', 'Locationabbr', 'Question', 'Response', 'Break_Out', 'Break_Out_Category', 'Sample_Size', 'Data_value', 'QuestionID', 'ResponseID']]

    grouped_tobacco = tobacco_breakout.groupby(['Year', 'Response', 'Break_Out'])[['Sample_Size']].sum().reset_index()

    # Filter data for the specified year
    grouped_tobacco = grouped_tobacco[grouped_tobacco['Year'] == yr]

    # Set the order for 'Break_Out' and 'Response'
    if cat_order is not None:
        grouped_tobacco['Break_Out'] = pd.Categorical(grouped_tobacco['Break_Out'], categories=cat_order, ordered=True)
    if response_order is not None:
        grouped_tobacco['Response'] = pd.Categorical(grouped_tobacco['Response'], categories=response_order, ordered=True)

    # Sort values by 'Break_Out' and 'Response'
    sorted_tobacco = grouped_tobacco.sort_values(['Break_Out', 'Response'])
    df_pivot = sorted_tobacco.pivot(index='Break_Out', columns='Response', values='Sample_Size').fillna(0)

    # Plot total respondent chart
    ax = df_pivot.plot(kind='bar', stacked=True, figsize=(8, 6), color=colors)
    plt.title(f'{breakout_cat} Breakdown of US Tobacco Usage Trends in {yr}')
    plt.xlabel(f'{breakout_cat}')
    plt.ylabel('Respondents')
    plt.legend(title='Response', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
    
    # Plot proportional chart
    df_pivot_normalized = df_pivot.div(df_pivot.sum(axis=1), axis=0) * 100
    ax = df_pivot_normalized.plot(kind='bar', stacked=True, figsize=(8, 6), color=colors)
    plt.title(f'PROPORTIONAL {breakout_cat} Breakdown of US Tobacco Usage Trends in {yr}')
    plt.xlabel(f'{breakout_cat}')
    plt.ylabel('Percentage of Respondents')
    plt.legend(title='Response', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```
