```python
import pygris
import geopandas as gpd
```

```python
from pygris.data import get_census


#total is 389,000 entries
dt_2022  = get_census(dataset = "acs/acs5",
                           variables = ["B07013_002E","B07013_003E"],
                           year = 2022,
                           params = {
                             "for": "tract:*",
                             "in": "state:10"},
                           guess_dtypes = True,
                           return_geoid = True
)

dt_2022 = dt_2022.rename(columns={"B07013_002E": "Owner Occupied 2022", "B07013_003E": "Renter Occupied 2022"})
dt_2022["Total Occupied Housing Units 2022"] = dt_2022["Owner Occupied 2022"] + dt_2022["Renter Occupied 2022"]

dt_2022.head()
```

```python
dt_2022["Percent Owner Occupied 2022"] = dt_2022["Owner Occupied 2022"] / dt_2022["Total Occupied Housing Units 2022"]
dt_2022.head()
```

```python
from pygris import tracts
import geopandas as gpd
import matplotlib.pyplot as plt

# Download Delaware shapefile
de_pytracts = tracts(state="DE", year=2022)

gdf = de_pytracts.merge(right=dt_2022, how="left", left_on="GEOID", right_on="GEOID")

gdf.plot(column="Percent Owner Occupied 2022", legend=True, cmap='OrRd', figsize=(10, 10))
plt.title('Percent Owner Occupied Housing Units DE (2022)')
plt.show()
```

```python
dt_2012  = get_census(dataset = "acs/acs5",
                           variables = ["B07013_002E","B07013_003E"],
                           year = 2012,
                           params = {
                             "for": "tract:*",
                             "in": "state:10"},
                           guess_dtypes = True,
                           return_geoid = True
)

dt_2012 = dt_2012.rename(columns={"B07013_002E": "Owner Occupied 2012", "B07013_003E": "Renter Occupied 2012"})
dt_2012["Total Occupied Housing Units 2012"] = dt_2012["Owner Occupied 2012"] + dt_2012["Renter Occupied 2012"]
dt_2012["Percent Owner Occupied 2012"] = dt_2012["Owner Occupied 2012"] / dt_2012["Total Occupied Housing Units 2012"]
dt_2012.head()
```

```python
from pygris import tracts
import geopandas as gpd
import matplotlib.pyplot as plt

# Download Delaware shapefile
de_pytracts = tracts(state="DE", year=2012)

gdf = de_pytracts.merge(right=dt_2012, how="left", left_on="GEOID", right_on="GEOID")

gdf.plot(column="Percent Owner Occupied 2012", legend=True, cmap='OrRd', figsize=(10, 10))
plt.title('Percent Owner Occupied Housing Units DE (2012)')
plt.show()
```

```python
# Merge the 2012 and 2022 data on GEOID
merged_data = dt_2012.merge(dt_2022, on="GEOID")

# Calculate the percentage of owner-occupied units for each year
merged_data["Pct Owner Occupied 2012"] = merged_data["Owner Occupied 2012"] / merged_data["Total Occupied Housing Units 2012"] * 100
merged_data["Pct Owner Occupied 2022"] = merged_data["Owner Occupied 2022"] / merged_data["Total Occupied Housing Units 2022"] * 100

# Calculate the percent change from 2012 to 2022
merged_data["Pct Change Owner Occupied"] = (merged_data["Pct Owner Occupied 2022"] - merged_data["Pct Owner Occupied 2012"])
merged_data.head()
```

```python
de_pytracts_2022 = tracts(state="DE", year=2022)

gdf = de_pytracts_2022.merge(right=merged_data, how="left", left_on="GEOID", right_on="GEOID")

gdf.plot(column="Pct Change Owner Occupied", edgecolor="black", legend=True,  missing_kwds={"color":"white"}, cmap='RdBu', figsize=(10, 10))
plt.title('Percent Change In Owner Occupied Housing Units DE (2012-2022)')
plt.show()
```
